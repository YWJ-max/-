# 多线程多进程高并发网络爬虫管理算法

```python
import Get_Data
import multiprocessing
import threading
import requests

class Multi_Processing_Threading:
    def cheef(start,stop):
        for number in range(start,stop):
            try:
                Get_Data.main(number)
            except Exception as ex:
                print(ex)
                pass

    def Multi_Threading(mini_num, max_num,th_num):
        difference = max_num - mini_num  # 相差
        remainder = difference % th_num  # 余数
        interval = int((difference - remainder) / th_num)  # 相差的数
        Th_list = []
        for number in range(0, th_num):
            Th_list.append(tuple(((number * interval) + mini_num, (number + 1) * interval + mini_num)))
        if remainder == 0:
            for number in Th_list:
                #print('执行范围', number[0], number[1])
                Th = threading.Thread(target=Multi_Processing_Threading.cheef, kwargs={"start": number[0], "stop": number[1]})
                # Th2 = threading.Thread(target=cheef,kwargs={"start":50,"stop":100})
                Th.start()
        else:
            last_tuple = tuple((th_num * interval + mini_num, max_num))
            Th_list.append(last_tuple)
            #print(Th_list)
            for number in range(0, len(Th_list) - 1):
                #print(Th_list[number])
                Th = threading.Thread(target=Multi_Processing_Threading.cheef,kwargs={"start":Th_list[number][0],"stop":Th_list[number][1]})
                Th.start()
            Th = threading.Thread(target=Multi_Processing_Threading.cheef, kwargs={"start": Th_list[-1][0], "stop": Th_list[-1][1]})
            Th.start()

    def Multi_Processing(mini_num, max_num,mu_num,th_num):
        difference = max_num - mini_num  # 相差
        remainder = difference % mu_num  # 余数
        interval = int((difference - remainder) / mu_num)  # 相差的数
        Th_list = []
        for number in range(0, mu_num):
            Th_list.append(tuple(((number * interval) + mini_num, (number + 1) * interval + mini_num)))
        if remainder == 0:
            for number in Th_list:
                #print('执行范围', number[0], number[1])
                Th = multiprocessing.Process(target=Multi_Processing_Threading.Multi_Threading, kwargs={"mini_num": number[0], "max_num": number[1],"th_num":th_num})
                # Th2 = threading.Thread(target=cheef,kwargs={"start":50,"stop":100})
                Th.start()
        else:
            last_tuple = tuple((mu_num * interval + mini_num, max_num))
            Th_list.append(last_tuple)
            #print(Th_list)
            for number in range(0, len(Th_list) - 1):
                #print(Th_list[number])
                Th = multiprocessing.Process(target=Multi_Processing_Threading.Multi_Threading,kwargs={"mini_num":Th_list[number][0],"max_num":Th_list[number][1],"th_num":th_num})
                Th.start()
            Th = multiprocessing.Process(target=Multi_Processing_Threading.Multi_Threading, kwargs={"mini_num": Th_list[-1][0], "max_num": Th_list[-1][1],"th_num":th_num})
            Th.start()

if __name__=='__main__':
    #第一个参数是起始值，第二是终止值，第三是进程数，第四是进程中的最大线程数
    Multi_Processing_Threading.Multi_Processing(1,40000,5,5)
```

# 多线程多进程高并发分布式网络爬虫管理算法

```python
import Get_Data_Agency
import multiprocessing
import threading
import requests
import os

class VPN:
    def gain_data():  # 获取代理IP和端口
        try:
            data_dict = {}
            # res = requests.get('https://ip.jiangxianli.com/api/proxy_ips?country=美国')
            res = requests.get('https://niubility666.cn/china.html')
            res.encoding = 'utf-8'
            null = ''
            data = eval(res.text)
            ip_data = data['data']['data']
            for i in range(0, len(ip_data)):
                ip = ip_data[i]['ip']
                port = ip_data[i]['port']
                data_dict[ip] = port
            # print(data_dict)
            return data_dict
        except Exception as ex:
            print(ex)
            pass

    def connetction():  # 测试接口可用性，并返回可用接口
        try:
            data = VPN.gain_data()  # 获取IP和端口
            available_list = []
            for key in data:
                try:
                    url = "http://pv.sohu.com/cityjson?ie=utf-8"
                    ip = str(key)
                    port = str(data[key])
                    new_data = {
                        "http":'http://' +ip + ":" + port
                    }
                    print("代理后的IP: " + requests.get(url, proxies=new_data,timeout=4).text)
                    available_list.append(new_data)

                except Exception as ex:
                    print('a',ex)
                    pass
            print("切换回自己的IP: " + requests.get(url, proxies={"http": ""}).text)  # 结束后切回自己的IP地址
            return available_list  # 返回可用代理的数组（new_data）
        except Exception as ex:
            print(ex)
            pass



class Multi_Processing_Threading_Distributed:
    def __init__(self,init_Min,init_Max,init_Multi):
        self.Min = init_Min
        self.Max = init_Max
        self.Multi = init_Multi

    def cheef(start,stop,agent):
        for number in range(start, stop):
            Get_Data_Agency.main(number,agent)


    def Multi_Threading(mini_num, max_num):
        proxies_data = VPN.connetction()
        th_num = len(proxies_data)
        difference = max_num - mini_num  # 相差
        remainder = difference % th_num  # 余数
        interval = int((difference - remainder) / th_num)  # 相差的数
        Th_list = []
        for number in range(0, th_num):
            Th_list.append(
                tuple(((number * interval) + mini_num, (number + 1) * interval + mini_num, proxies_data[number])))
        if remainder == 0:
            for number in Th_list:
                Th = threading.Thread(target=Multi_Processing_Threading_Distributed.cheef, kwargs={"start": number[0], "stop": number[1], "agent": number[2]})
                Th.start()
        else:
            last_tuple = tuple((th_num * interval + mini_num, max_num))
            Th_list.append(last_tuple)
            for number in range(0, len(Th_list) - 1):
                Th = threading.Thread(target=Multi_Processing_Threading_Distributed.cheef, kwargs={"start": Th_list[number][0], "stop": Th_list[number][1],"agent": Th_list[number][2]})
                Th.start()
            Th = threading.Thread(target=Multi_Processing_Threading_Distributed.cheef,kwargs={"start": Th_list[-1][0], "stop": Th_list[-1][1], "agent": Th_list[number][2]})
            Th.start()

    def Multi_Processing(mini_num, max_num,mu_num):
        difference = max_num - mini_num  # 相差
        remainder = difference % mu_num  # 余数
        interval = int((difference - remainder) / mu_num)  # 相差的数
        Th_list = []
        for number in range(0, mu_num):
            Th_list.append(tuple(((number * interval) + mini_num, (number + 1) * interval + mini_num)))
        if remainder == 0:
            for number in Th_list:
                Th = multiprocessing.Process(target=Multi_Processing_Threading_Distributed.Multi_Threading, kwargs={"mini_num": number[0], "max_num": number[1]})
                Th.start()
        else:
            last_tuple = tuple((mu_num * interval + mini_num, max_num))
            Th_list.append(last_tuple)
            for number in range(0, len(Th_list) - 1):
                Th = multiprocessing.Process(target=Multi_Processing_Threading_Distributed.Multi_Threading,kwargs={"mini_num":Th_list[number][0],"max_num":Th_list[number][1]})
                Th.start()
            Th = multiprocessing.Process(target=Multi_Processing_Threading_Distributed.Multi_Threading, kwargs={"mini_num": Th_list[-1][0], "max_num": Th_list[-1][1]})
            Th.start()

    def main(self):
        VPN.connetction()
        Multi_Processing_Threading_Distributed.Multi_Processing(self.Min,self.Max,self.Multi)

if __name__=='__main__':
    #第一个参数是起始值，第二是终止值，第三是进程数
    p = Multi_Processing_Threading_Distributed(1,40000,5)
    p.main()
    
```
